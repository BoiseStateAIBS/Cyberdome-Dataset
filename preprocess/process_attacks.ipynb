```python
flattened_data = []

for day in attacks_data:
    day_label = day["day"]
    for attack in day["attacks"]:
        # Extracting and handling multiple timestamps and multiple targets
        timestamps = attack.get("timestamps", [attack.get("timestamp")])
        targets = attack.get("targets", [attack.get("target")])
        
        for timestamp in timestamps:
            for target in targets:
                flattened_data.append({
                    "Day": day_label,
                    "Attack": attack["attack"],
                    "ATT&CK_ID": attack["ATT&CK_ID"],
                    "Target": target,
                    "Timestamp": timestamp
                })


# Convert to DataFrame
attacks_df = pd.DataFrame(flattened_data)
```

```python

def split_timestamp_range(row):
    if "-" in row['Timestamp']:
        # Parse the date, start time, and end time from the 'Timestamp' column
        date_part, time_range, am_pm, tz = row['Timestamp'].split()
        start_time, end_time = time_range.split('-')
        
        # Create new rows for start and end times with "continued" type
        start_timestamp = f"{date_part} {start_time} {am_pm} {tz}"
        end_timestamp = f"{date_part} {end_time} {am_pm} {tz}"
        
        return [
            {**row, 'Timestamp': start_timestamp, 'Type': 'continued'},
            {**row, 'Timestamp': end_timestamp, 'Type': 'continued'}
        ]
    else:
        # For unique timestamps, just return the row with "unique" type
        return [{**row, 'Type': 'unique'}]

# Apply the function to each row and expand the DataFrame
attacks_df = pd.DataFrame([new_row for index, row in attacks_df.iterrows() for new_row in split_timestamp_range(row)])

```
